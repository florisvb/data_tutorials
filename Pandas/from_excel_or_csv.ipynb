```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pandas.read_excel('data.xlsx') # similar command for CVS file
```

```python
df
```

```python
# plot something simple
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(df.time, df['x_value'])
```

```python
# choose rows where 'label' column is not NAN
dfq = df[df.label.notna()] 

fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(dfq.x_value)
```

```python
# choose rows where 'x_value' column is > 1.5, using query
query = 'x_value > ' + str(1.5) 
dfq = df.query(query)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(dfq.x_value)
```
